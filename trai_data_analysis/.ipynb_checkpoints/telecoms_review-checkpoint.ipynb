*This is the list of cellular operator names according to data file.*

```python
operator_names = pd.unique(trai_data.operator)
```

```python
operator_names
```

*Now we want to see the location of each of the operators. We will 
do this by using a 2-d plot.*

*Processing the data so that only the relevant data of coordinates in India is there.*

```python
# The variable trai_data_processed is used whenever the data is filtered toemporarily to suit our purpose
trai_data_processed = trai_data[trai_data.latitude > 0 ]
trai_data_processed = trai_data_processed[trai_data_processed.longitude > 0]
trai_data_processed = trai_data_processed[trai_data_processed.longitude < 95]
```

```python
trai_data_processed.count()
```

*Plotting the graph keeping each operator name as label. Refer to the pdf file created.(Zoom in for more clarity) *

```python
plt.figure(figsize = (17,20))
for op in operator_names:
    ix = trai_data_processed.operator == op
    x, y = trai_data_processed.longitude[ix], trai_data_processed.latitude[ix]
    plt.plot(x, y, ".", label = op)
plt.title("Locations of different tele-companies in operation - May 2018")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(loc = "lower right")
plt.savefig("Operator_location.pdf")
```

*Now this is histogram time!*

Histogram of the rating column overall and divided according to companies.

```python
trai_data.rating.plot(kind = 'hist', range = [0,5])
plt.title("Rating of all users")
plt.xlabel("Rating")
plt.savefig("Rating_all_hist.pdf")
```

```python
for op in operator_names:
    ix = trai_data_processed.operator == op
    trai_data_processed.rating[ix].plot(kind = 'hist', range = [1,5], legend = True, label = op)
plt.title("Rating of users divided according to Telecom Cos")
plt.xlabel("Rating")
plt.savefig("Rating_teleco_hist.pdf")
```

Histogram of the rating column overall and divided according to companies.

```python
state_names = pd.unique(trai_data_processed.state_name) 
for state in state_names:
    ix = trai_data_processed.state_name == state
    trai_data_processed.rating[ix].plot(kind = 'hist', range = [1,5], legend = True, label = state, figsize =[10,20])
plt.title("Rating of users divided according to States")
plt.xlabel("Rating")
plt.savefig("Rating_state_hist.pdf")
```
